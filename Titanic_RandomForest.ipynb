```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base_treino = pd.read_csv('train.csv')
base_treino
```

```python
base_teste = pd.read_csv('test.csv')
base_teste
```

```python
base_completa = pd.concat([base_treino, base_teste], axis = 0)
base_completa
```

```python
base_completa.describe()
```

```python
base_completa['Cabin'].value_counts()
```

```python
base_completa['Ticket'].value_counts()
```

```python
base_completa.drop(['Name', 'Ticket', 'Cabin'], axis = 1, inplace = True)
base_completa
```

```python
base_completa.isnull().sum()
```

```python
base_completa['Age'].fillna(base_completa['Age'].mean(), inplace = True)
base_completa
```

```python
base_completa['Embarked'].value_counts()
```

```python
base_completa['Embarked'].fillna('S', inplace = True)
base_completa
```

```python
base_completa['Fare'].fillna(base_completa['Fare'].mean(), inplace = True)
base_completa
```

```python
base_completa.isnull().sum()
```

```python
sns.countplot(x = 'Survived', data = base_completa)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
base_completa['Sex'] = label_encoder.fit_transform(base_completa['Sex'])
base_completa['Embarked'] = label_encoder.fit_transform(base_completa['Embarked'])
base_completa
```
